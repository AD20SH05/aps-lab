# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def dijkstra(graph, start):

    distances = {}
    visited = []
    previous = {}

    # Initialize distances
    for node in graph:
        distances[node] = float('inf')
        previous[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        # Find node with minimum distance
        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.append(min_node)

        # Update neighbor distances
        for neighbor, weight in graph[min_node].items():

            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous[neighbor] = min_node

    return distances, previous


# Function to display shortest path
def get_path(previous, start, target):

    path = []

    while target is not None:
        path.append(target)
        target = previous[target]

    path.reverse()

    if path and path[0] == start:
        return path
    return []


# User Input Graph
graph = {}

while True:
    try:
        n = int(input("Enter number of nodes: "))
        if n <= 0:
            print("Number of nodes must be positive. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter an integer for the number of nodes.")

for i in range(n):
    node = input(f"\nEnter node name {i+1}/{n}: ")
    graph[node] = {}

    while True:
        try:
            neighbors_count = int(input(f"Enter number of neighbors for {node}: "))
            if neighbors_count < 0:
                print("Number of neighbors cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer for the number of neighbors.")

    for j in range(neighbors_count):
        neighbor = input(f"Enter neighbor node {j+1}/{neighbors_count} for {node}: ")
        while True:
            try:
                weight = int(input(f"Enter edge weight for {node} to {neighbor}: "))
                if weight < 0:
                    print("Weight cannot be negative. Please try again.")
                    continue
                break
            except ValueError:
                print("Invalid input. Please enter an integer for the edge weight.")

        graph[node][neighbor] = weight


# Test with different source nodes
while True:
    source = input("\nEnter source node: ")

    if source not in graph:
        print("Invalid source node. Please choose from existing nodes.")
        continue

    distances, previous = dijkstra(graph, source)

    print("\nShortest Distances and Paths:\n")

    for node in graph:
        path = get_path(previous, source, node)

        print(f"{source} -> {node}")
        print("Distance:", distances[node])
        if path:
            print("Path:", " -> ".join(path))
        else:
            print("Path: No path found")
        print()

    choice = input("Test another source node? (yes/no): ")

    if choice.lower() != "yes":
        break
```
